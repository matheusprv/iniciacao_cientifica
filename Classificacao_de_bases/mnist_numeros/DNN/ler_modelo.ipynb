```python
import load_dataset as ld

import tensorflow as tf
from tensorflow import keras
import h5py
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
x_test, y_test = ld.load_dataset("../mnist_test.csv", 28, 28)
```

```python
model = ld.load_model("model.json", "model.h5")
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=1) 
print('Test accuracy:', test_acc)
```

```python
def print_image(image):
    plt.figure()
    plt.imshow(image[0])
    plt.show
    plt.pause(0.001)
```

```python
while True:
    index = int(input("Digite o index desejado: "))

    index -= 1

    image = x_test[index].reshape((1,28,28))

    print_image(image)

    prediction = model.predict(image)

    print("\nResultado encontrado: ",np.argmax(prediction))
    print("Resultado esperado: ", int(y_test[index]))
```
