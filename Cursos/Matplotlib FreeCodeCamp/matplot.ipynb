```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic graph

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)

font_info_dict = {
    'fontname': "Arial", 
    'fontsize': 20
}

plt.title("Title of the Graph", fontdict = font_info_dict)
plt.xlabel("X axis title")
plt.ylabel("Y axis title")

plt.show()
```

```python
x = np.array([0,1,2,3,4,5]) 
y = x * 2

plt.plot(x,y)

font_info_dict = {
    'fontname': "Arial", 
    'fontsize': 20
}

plt.title("Title of the Graph", fontdict = font_info_dict)
plt.xlabel("X axis title")
plt.ylabel("Y axis title")

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

```python
x = np.array([0,1,2,3,4,5]) 
y = x * 3

plt.plot(x, y, label = '2x', color = "red", linewidth = 1, marker = 'x', markersize = 10)
plt.plot((y*y), x, label = "second", color = "purple", linestyle = "dotted")

font_info_dict = {
    'fontname': "Arial", 
    'fontsize': 20
}

plt.title("Title of the Graph", fontdict = font_info_dict)
plt.xlabel("X axis title")
plt.ylabel("Y axis title")


plt.legend()

plt.show()
```

# Bar Chart

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

plt.bar(labels, values)

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[1].set_hatch("/")
bars[0].set_color("red")

plt.show()
```
